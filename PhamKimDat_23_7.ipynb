```python
import osmnx as ox
import networkx as nx
import folium
from geopy.geocoders import Nominatim
dc_ktx = "43 - 45 Nguyen Chi Thanh, Ho Chi Minh"
dc_nha = "124 Pham Phu Thu, Ho Chi Minh"
geolocator = Nominatim(user_agent = "geoapi", timeout = 100)
td_ktx = geolocator.geocode(dc_ktx)
td_nha = geolocator.geocode(dc_nha)
ktx = [td_ktx.latitude, td_ktx.longitude]
nha = [td_nha.latitude, td_nha.longitude]
G = ox.graph_from_point(nha, dist = 10000, network_type = "drive")
start_point = ox.distance.nearest_nodes(G, ktx[1],ktx[0])
end_point = ox.distance.nearest_nodes(G, nha[1],nha[0])
route_dijkstra = nx.shortest_path(G, start_point, end_point, weight = "length", method = "dijkstra")

m = folium.Map(location = [td_ktx.latitude,td_ktx.longitude], zoom_start=13)

route_coords_dijkstra = []
for node in route_dijkstra:
  lat = G.nodes[node]['y']
  lon = G.nodes[node]['x']
  route_coords_dijkstra.append((lat,lon))
folium.PolyLine(route_coords_dijkstra, color = "blue", weight = 3, opacity = 0.5, tooltip = "Dijkstra Route").add_to(m)


folium.Marker(ktx, popup = "Điểm đón").add_to(m)
folium.Marker(nha, popup = "Điểm đến").add_to(m)
m
```
