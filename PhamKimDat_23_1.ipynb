```python
import folium
#UEH
m = folium.Map(location=[10.7610585,106.6657831], zoom_start = 13)
folium.Marker([10.7610585,106.6657831], popup="UEH cơ sở B", tooltip="Click vào", icon=folium.Icon(color="red")).add_to(m)
#Địa điểm công cộng
place = [
    ("Bệnh viện Nguyễn Tri Phương", 10.7550078,106.6679596),
    ("Vòng xoay Nguyễn Tri Phương", 10.7598698,106.6663278),
    ("Nhà thờ thánh nữ Jeanne d'Arc", 10.7569302,106.6646537),
    ("Công viên Văn Lang", 10.7564807,106.6651499),
    ("Bệnh viện Hòa Hảo", 10.7623879,106.6677098)]
for name, lat, lon in place:
  folium.Marker(
      location = [lat, lon],
      popup = f"{name}"
  ).add_to(m)
m
```
