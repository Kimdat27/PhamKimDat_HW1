```python
import folium
import random
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapi",timeout = 100)
center = geolocator.geocode("Ben Thanh Market, Ho Chi Minh City")
td = [center.latitude, center.longitude]
m = folium.Map(location=td, zoom_start = 13)
folium.Marker(td, popup = "Ben Thanh Market", icon=folium.Icon(color = "red")).add_to(m)
data=[]
for i in range(10):
  lat = center.latitude + random.uniform(0.001,0.01)
  lon = center.longitude + random.uniform(0.001,0.01)
  w = random.randint(1,10)
  data.append([lat, lon, w])

for i in range(30):
  lat1 = center.latitude - random.uniform(0.001,0.01)
  lon1 = center.longitude - random.uniform(0.001,0.01)
  w1 = random.randint(1,10)
  data.append([lat1, lon1, w1])
HeatMap(data).add_to(m)
m
```
