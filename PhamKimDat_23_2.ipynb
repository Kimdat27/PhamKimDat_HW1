```python
import folium
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
geolocator = Nominatim(user_agent = "geoapi", timeout = 100)
place = [
    "Ben Thanh Market, Ho Chi Minh City",
    "Ba Chieu Market, Ho Chi Minh City",
    "Ho Thi Ki Market, Ho Chi Minh City",
    "Nhat Tao Market, Ho Chi Minh City",
    "Tan Son Nhat Airport, Ho Chi Minh City",
    "Van Lang park, Ho Chi Minh City",
    "Van Hanh Mall, Ho Chi Minh City",
    "Dam Sen park, Ho Chi Minh City",
    "Suoi Tien park, Ho Chi Minh City",
    "30/4 Hospital, Ho Chi Minh City"]
#Diem trung tam
center = geolocator.geocode("Dinh doc lap, Ho Chi Minh City")
center_coord = [center.latitude,center.longitude]
m = folium.Map(location = center_coord, zoom_start = 13)
folium.Marker(center_coord, popup = "Independence Palce", icon = folium.Icon(color = "red")).add_to(m)
for name in place:
  loc = geolocator.geocode(name)
  if loc:
    td = [loc.latitude, loc.longitude]
    distance = geodesic(center_coord,td).km
    folium.Marker(
        location = td,
        popup = f"{name} to Independence Palace: {distance:.2f}km"
        ).add_to(m)
m
```
